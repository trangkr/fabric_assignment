```python
%%sql
CREATE OR REPLACE TABLE stg_customers AS
SELECT DISTINCT
    customer_id,
    TRIM(customer_name)     AS customer_name,
    UPPER(TRIM(city))       AS city,
    TRIM(segment)           AS segment,
    CAST(last_modified AS DATE) AS last_modified
FROM raw_customers
WHERE customer_id IS NOT NULL
```

```python
%%sql
CREATE OR REPLACE TABLE stg_products AS
SELECT DISTINCT
    product_id,
    TRIM(product_name)          AS product_name,
    TRIM(category)              AS category,
    CAST(unit_price AS DOUBLE)  AS unit_price,
    CAST(last_modified AS DATE) AS last_modified
FROM raw_products
WHERE product_id  IS NOT NULL
AND   CAST(unit_price AS DOUBLE) > 0
```

```python
%%sql
CREATE OR REPLACE TABLE stg_sales AS
SELECT DISTINCT
    order_id,
    CAST(order_date    AS DATE)   AS order_date,
    customer_id,
    product_id,
    CAST(quantity      AS INT)    AS quantity,
    CAST(line_amount   AS DOUBLE) AS line_amount,
    CAST(last_modified AS DATE)   AS last_modified
FROM raw_sales
WHERE order_id    IS NOT NULL
AND   customer_id IS NOT NULL
AND   product_id  IS NOT NULL
AND   CAST(quantity AS INT) > 0
AND   CAST(line_amount AS DOUBLE) > 0
```

```python
%%sql
SELECT 'stg_customers' AS tbl, COUNT(*) AS rows FROM stg_customers
UNION ALL
SELECT 'stg_products',         COUNT(*)          FROM stg_products
UNION ALL
SELECT 'stg_sales',            COUNT(*)          FROM stg_sales
```
