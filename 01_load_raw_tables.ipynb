```python
# Welcome to your new notebook
# Type here in the cell editor to add code!
# Load CSV vào Delta Table (raw zone)
df_customers = spark.read.option("header", "true").option("inferSchema", "true") \
    .csv("Files/raw_input/customers.csv")

df_customers.write.mode("overwrite").saveAsTable("raw_customers")
print(f"Loaded {df_customers.count()} rows into raw_customers")
df_customers.show()
```

```python
df_products = spark.read.option("header", "true").option("inferSchema", "true") \
    .csv("Files/raw_input/products.csv")

df_products.write.mode("overwrite").saveAsTable("raw_products")
print(f"Loaded {df_products.count()} rows into raw_products")
df_products.show()
```
